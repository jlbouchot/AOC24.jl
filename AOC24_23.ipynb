```julia
function BronKerboschAlgo(R, P, X, neighbours)
    # println("R = ", R)
    # println("P = ", P)
    # println("X = ", X)
    
    if length(P) == 0 && length(X) == 0 # Nothing left to do!
        # println("Found a clique: ", R)
        return R
    end
    # Not pivoting
    # println("Setting up solutions")
    solutions = Set{String}()
    for v in P
        # s = 
        
        s = BronKerboschAlgo(union(R,[v]), intersect(P, Set(neighbours[v])), intersect(X, Set(neighbours[v])), neighbours)
        # println("s is ", s, " while solutions is ", solutions)
        if length(s) > length(solutions)
            solutions = s
        end
        # println("solutions are ", s, " of length ", length(s))
        # for s1 in s
        #     push!(solutions, [s])
        # end
        delete!(P, v)
        push!(X,v)
    end
    return solutions
end
```
