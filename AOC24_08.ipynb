```julia
tot_sum = open("08_1.txt") do f
    tot = 0
    antennas = DefaultDict{Char, Array{Array{Int,1},1}}(() -> Array{Array{Int,1},1}())
    all_antennas = Set()
    antinodes = Set()
    i = 1
    maxi = 0
    maxj = 0
    for l in eachline(f)
        j = 1
        for c in l
            if c != '.'
                push!(all_antennas, [i,j])
                push!(antennas[c], [i,j])
            end
            j +=1
            maxj = j
        end
        i +=1
        maxi = i
    end

    for k in keys(antennas)
        sort!(antennas[k], by = x -> x[1])
        nb_antennas = length(antennas[k])
        for i in range(1,nb_antennas-1)
            a1 = antennas[k][i]
            for j in range(i+1, nb_antennas)
                a2 = antennas[k][j]
                deltai = a1[1] - a2[1]
                deltaj = a1[2] - a2[2]
                new_element = a2 - [deltai, deltaj]
                if (0 < new_element[1] < maxi) && (0 < new_element[2] < maxj)
                    if !(new_element in antinodes)
                        push!(antinodes, new_element)
                        tot += 1
                    end
                end
                new_element = a1 + [deltai, deltaj]
                if (0 < new_element[1] < maxi) && (0 < new_element[2] < maxj)
                    if !(new_element in antinodes)
                        push!(antinodes, new_element)
                        tot += 1
                    end
                end
            end
        end
    end
    (tot)
end
```
