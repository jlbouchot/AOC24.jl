```julia
reports = Dict()

f = read("03_1.txt", String)
i = 1
tot_sum = 0
all_values = getfield.(eachmatch(r"mul\(([0-9]{1,3},[0-9]{1,3}\))", f), :match)
for a_v in all_values
    
    d1,d2 = getfield.(eachmatch(r"[0-9]{1,3}", a_v), :match)
    cur_val = parse(Int,d1)*parse(Int,d2)
    tot_sum += cur_val
    println(a_v, " should multiply ", d1, " with ", d2, " = ", cur_val, " tot = ", tot_sum)
    
end
println(tot_sum)
```
